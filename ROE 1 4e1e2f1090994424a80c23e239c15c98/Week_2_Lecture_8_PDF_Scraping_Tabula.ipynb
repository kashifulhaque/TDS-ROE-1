```python
os.chdir('/content/drive/Shareddrives/Tools for Data Science/India Election Data/')
results1 = {}
results2 = {}
i = 1
for filename in fnames:
  try:

    df = tabula.read_pdf(filename,pages='1-100',multiple_tables=True, silent=True)
    temp = df[19].iloc[:,-1:].dropna()
    temp.columns = ['votes']
    temp = temp[~temp['votes'].isin(['Total','Valid','Votes','votes','TOTAL','VALID','VOTES','total','valid'])]
    temp['votes'] = temp['votes'].astype(int)
    answer1 = len(temp[temp['votes']<100000])
    answer2 = temp['votes'].mean()
    print(i, filename, len(temp), answer1, answer2)
    i = i+1
    results1[filename] = answer1
    results2[filename] = answer2
  except:
    print('Skipping file')
    i = i+1
    continue
    
df1 = pd.DataFrame(results1.items(), columns=['filename', 'answer1'])
df2 = pd.DataFrame(results2.items(), columns=['filename', 'answer2'])
df = pd.merge(df1,df2,on=['filename'],how='inner')
df.to_csv('results.csv')
```

```python
# Save contents from url into folder_location
url = 'https://www.premierleague.com/publications'
folder_location = r'/content/drive/MyDrive/Colab Notebooks/premier_league'
if not os.path.exists(folder_location):
    os.mkdir(folder_location)
```

Actual Code

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")     

# Loop through all PDF links in the page
for link in soup.select("a[href$='.pdf']"):
    # Local lile name is the same as PDF file name in the URL (ignoring rest of the path)
    # https://premierleague-static-files.s3.amazonaws.com/premierleague/document/2016/07/02/e1648e96-4eeb-456e-8ce0-d937d2bc7649/2011-12-premier-league-season-review.pdf
    filename = os.path.join(folder_location, link['href'].split('/')[-1])
    with open(filename, 'wb') as f:
        f.write(requests.get(urljoin(url,link['href'])).content)
```
